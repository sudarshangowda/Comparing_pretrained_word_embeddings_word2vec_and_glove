```
import keras
import tensorflow as tf
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import *
from keras.layers import *
from keras.callbacks import *
```

```
train =pd.read_csv('/Train.csv', encoding = "utf8")
```

```
valid =pd.read_csv('/Valid.csv', encoding = "utf8")
```

```
x_tr, y_tr = train['text'].values, train['label'].values
x_val, y_val = valid['text'].values, valid['label'].values
```

```
tokenizer = Tokenizer()
```

```
t = tokenizer.fit_on_texts(list(x_tr))
```

```
x_tr_se = tokenizer.texts_to_sequences(x_tr)
x_val_se = tokenizer.texts_to_sequences(x_val)
```

```
x_tr_seq= pad_sequences(x_tr_se, maxlen=100)
x_val_seq= pad_sequences(x_val_se, maxlen=100)
```

```
model = Sequential()
```

```
text_lines = list()
lines = df['text'].values.tolist()
```

```
for line in lines:
  tokens = word_tokenize(str(line))
  tokens = [w.lower() for w in tokens]
  table = str.maketrans('', '', string.punctuation)
  stripped = [w.translate(table) for w in tokens]
  words = [word for word in stripped if word.isalpha()]
  stop_words =set(stopwords.words('english'))
  words =[w for w in words if not w in stop_words]
  text_lines.append(words)
```
